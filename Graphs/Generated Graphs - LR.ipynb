```python
import pandas
import matplotlib.pyplot as plt

train_df=pandas.read_json('train.json')

ax = train_df['cuisine'].value_counts().plot(kind='bar', title ="Cuisine types", figsize=(8, 3), legend=True, fontsize=12)
ax.set_ylabel("recipes", fontsize=12)
plt.show()
```

```python
%config IPCompleter.greedy=True

import pandas
import matplotlib.pyplot as plt
import seaborn as sns

train_df = pandas.read_csv('logisticRegression1.csv')

cuisine_count  = train_df['cuisine'].value_counts()
plt.figure(figsize=(20,10))
sns.barplot(cuisine_count.values, cuisine_count.index, alpha=1.0)
plt.title('logisticRegression1', fontsize=18)
plt.ylabel('Name of cuisine', fontsize=14)
plt.xlabel('# of recipes', fontsize=14)
plt.show()
```
